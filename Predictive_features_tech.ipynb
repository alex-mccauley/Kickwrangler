```python
today = pd.tslib.Timestamp('2015-6-13',tz=None)
one_year = pd.tslib.Timedelta(1,'Y')
def map_date(date):
    if date > today:
        return date - one_year
    else:
        return date
```

```python
# Simple method: projects with > 20 comments, look at last 4 months of comments
total_sentiment = sentiment_df.copy(deep=True)
total_comments = total_sentiment.groupby('project_id').count().reset_index()
total_comments.rename(columns={'comment_id':'total_comments'},inplace=True)
total_sentiment = pd.merge(total_sentiment, total_comments[['project_id','total_comments']],on='project_id',how='inner')

# > 20 comments
total_sentiment = total_sentiment[total_sentiment['total_comments']>20]

# dates past today are probably from 2014 - clean up
total_sentiment['date'] = total_sentiment['date'].apply(map_date)

# determine date of last comment
grouped = total_sentiment.groupby('project_id')
last_date = grouped.agg(np.max).reset_index()[['project_id','date']]
last_date.rename(columns={'date':'end_date'},inplace=True)
total_sentiment = pd.merge(total_sentiment, last_date,on='project_id',how='inner')

# remove comments 6 months before last comment
total_sentiment['date_from_end'] = total_sentiment['end_date'] - total_sentiment['date']
six_months = pd.tslib.Timedelta(6,'M')
total_sentiment = total_sentiment[(total_sentiment['date'] + six_months) > total_sentiment['end_date']]

# re-do total comments
total_comments = total_sentiment.groupby('project_id').count().reset_index()
total_sentiment.drop(axis=1,labels='total_comments',inplace=True)
total_sentiment = pd.merge(total_sentiment, total_comments[['project_id','total_comments']],on='project_id',how='inner')

# remove those with < 5 comments
total_sentiment = total_sentiment[total_sentiment['total_comments']>=5]

# compute average sentiment and pos/neg fraction
def is_pos(x):
    if x > 0:
        return 1
    else:
        return 0
def is_neg(x):
    if x < 0:
        return 1
    else:
        return 0

total_sentiment['sent_pos_avg'] = total_sentiment['sentiment'].apply(is_pos) / total_sentiment['total_comments']
total_sentiment['sent_neg_avg'] = total_sentiment['sentiment'].apply(is_neg) / total_sentiment['total_comments']
project_sentiments = total_sentiment.groupby('project_id').agg(np.sum).reset_index()[['project_id','sent_pos_avg','sent_neg_avg']]
project_sentiments['sent_avg']=project_sentiments['sent_pos_avg']-project_sentiments['sent_neg_avg']
project_sentiments['sent_frac'] = project_sentiments['sent_neg_avg']/(project_sentiments['sent_neg_avg']+project_sentiments['sent_pos_avg'])

project_sentiments = pd.merge(project_sentiments,total_comments[['project_id','total_comments']],on='project_id',how='left')
```
