```python
# total number of search results for projects
def get_number_of_hits(d):
    number_of_hits = driver.find_element_by_xpath('//*[@id="projects"]/div[2]/h3/b')
    return int(number_of_hits.text.split(' ')[0].replace(',',''))
```

```python
# get json directly:
# props to this site: http://www.kbremner.com/2014/03/15/kickstarter-part1.html
# example: 
#  https://www.kickstarter.com/discover/advanced.json?state=successful&category_id=16&woe_id=0&sort=newest&page=120
```

```python
from collections import defaultdict
import sys

def get_project_data(d, base_url):

    d.get(base_url.replace('.json', ''))
    number_of_hits = get_number_of_hits(d)
    print 'Extracting data for %d projects' % number_of_hits

    # determine number of pages
    results_per_page = 20
    number_of_pages = int(np.ceil(number_of_hits / results_per_page)) + 1 # add extra page for margin

#    number_of_pages = 50
    output_dict = defaultdict(dict)
    for page in range(number_of_pages):
        try:
            print 'Extracting page %d of %d' % (page, number_of_pages)
            sys.stdout.flush()

            page_url = base_url + '&page=%d' % page
            url_text = requests.get(page_url).text
            url_json = json.loads(url_text)

            for proj in url_json['projects']:
                # get project, assume 'id' is a unique key
                key = proj['id']
                output_dict.update({key:proj})
        except ValueError:
            print "Error on page %d" % page
    print 'Finished page %d of %d total' % (page, number_of_pages - 1)
    return output_dict                        
```

##Get user comments from each project

```python
def get_comments(d, base_url, delay = 1, max_comments=400):
    comments_url = base_url.replace('?ref=discovery','/comments')
    d.get(comments_url)
    
    # find total number of comments in this project
    total_comments_obj = d.find_element_by_xpath('//*[@id="comments_count"]')
    total_comments = int(total_comments_obj.text.split(' ')[-1].replace(',','')[1:-1])
    
    # estimate number of clicks
    comments_per_click = 50
    max_number_of_clicks = int(round(max_comments / comments_per_click))
    number_of_clicks = min([int(round(total_comments / comments_per_click) + 1), max_number_of_clicks])
        
    # click on button 
    if total_comments > 50:
        button = d.find_element_by_xpath('//*[@id="content-wrap"]/div[3]/section[6]/div/div/div/div[2]/div[2]/a')
        try:
            for ii in range(number_of_clicks):
                button.click()            
                time.sleep(delay)
        except:
            x=3

    comments = d.find_elements_by_class_name('NS_comments__comment')
    return comments, total_comments
```
